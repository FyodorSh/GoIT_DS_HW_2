12. Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

13. Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
salary_group_data = python_data.groupby('Должность').agg({'Зарплата.в.месяц' : ['min', 'max']})
salary_group_data
```

14. Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(row):
  return np.average([row['Зарплата.в.месяц']['min'], row['Зарплата.в.месяц']['max']])

salary_group_data['avg'] = salary_group_data.apply(fill_avg_salary, axis=1)
salary_group_data
```

15. Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
salary_descr = salary_group_data['avg'].describe()
```

```python
lang_stat = data.groupby('Язык.программирования').agg(lang_count = ('Язык.программирования', 'count'))
lang_stat = lang_stat.loc[lang_stat["lang_count"] >= lang_stat.mean()["lang_count"]].sort_values(by=["lang_count"], ascending=False)
plot = lang_stat.plot.pie(y='lang_count', figsize=(10, 10), legend=True)
```

```python
city_stat = data.groupby('Город').agg(city_count = ('Город', 'count'))
# city_stat
ax = city_stat.plot.barh(y='city_count', rot=0, figsize = (12,10))
```

```python
age_stat = pd.DataFrame({'lang':data['Язык.программирования'], 'age':data['Возраст']})
age_stat = age_stat.groupby('lang').agg(age_min = ('age','min'), age_mean = ('age','mean'),age_max = ('age','max'), age_count = ('age','count'))
age_stat = age_stat[lambda x: x['age_count'] > 40]
ax = age_stat.plot.bar(y=['age_min', 'age_mean', 'age_max'], rot=0, figsize=(20, 8))
```
